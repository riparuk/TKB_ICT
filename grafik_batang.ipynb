```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data yang telah disesuaikan
data = {
    "Bulan": ["Januari", "Januari", "Januari", "Februari", "Februari", "Februari"],
    "Kabupaten": ["Aceh Besar", "Aceh Barat", "Pidie", "Aceh Besar", "Aceh Barat", "Pidie"],
    "Pria": [10, 20, 20, 40, 20, 20],
    "Wanita": [25, 15, 20, 55, 10, 15],
}

df = pd.DataFrame(data)

# Mendapatkan daftar unik bulan dan kabupaten
bulan_unik = df["Bulan"].unique()
kabupaten_unik = df["Kabupaten"].unique()

# Membuat figure dan axis untuk dua subplot (satu untuk setiap bulan)
fig, axes = plt.subplots(nrows=1, ncols=len(bulan_unik), figsize=(12, 6), sharey=True)

# Jika hanya satu bulan, tetap buat dalam bentuk array
if len(bulan_unik) == 1:
    axes = [axes]

# Iterasi untuk setiap bulan
for i, bulan in enumerate(bulan_unik):
    ax = axes[i]  # Pilih subplot untuk bulan ini
    df_bulan = df[df["Bulan"] == bulan]  # Filter data berdasarkan bulan

    x = np.arange(len(kabupaten_unik))  # Posisi x untuk kabupaten
    bar_width = 0.4  # Lebar batang

    # Plot pria dan wanita secara berdampingan (side-by-side bars)
    ax.bar(x - bar_width / 2, df_bulan["Pria"], width=bar_width, label="Pria", color="blue")
    ax.bar(x + bar_width / 2, df_bulan["Wanita"], width=bar_width, label="Wanita", color="pink")

    # Set judul dan label sumbu
    ax.set_xticks(x)
    ax.set_xticklabels(kabupaten_unik, rotation=30)
    ax.set_title(f"Data Penduduk - {bulan}")
    ax.set_xlabel("Kabupaten")
    if i == 0:
        ax.set_ylabel("Jumlah Penduduk")  # Hanya di subplot pertama
    ax.legend()

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
